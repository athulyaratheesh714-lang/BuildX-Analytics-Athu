```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
df.isnull().sum()
```

```python
df["restaurant type"] = df["restaurant type"].fillna("Unknown")
df["cuisines type"] = df["cuisines type"].fillna("Unknown")
df["local address"] = df["local address"].fillna("Unknown")
```

```python
df["rate (out of 5)"] = df["rate (out of 5)"].fillna(df["rate (out of 5)"].mean())
```

```python
df = df.drop_duplicates()

print("Remaining rows:", len(df))
```

```python
text_columns = [
    "restaurant name",
    "restaurant type",
    "cuisines type",
    "area",
    "local address"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()

df["online_order"] = df["online_order"].astype(str).str.strip().str.lower()
df["table booking"] = df["table booking"].astype(str).str.strip().str.lower()
```

```python
df["rate (out of 5)"] = pd.to_numeric(df["rate (out of 5)"], errors="coerce")

df["num of ratings"] = pd.to_numeric(df["num of ratings"], errors="coerce")

df["avg cost (two people)"] = pd.to_numeric(df["avg cost (two people)"], errors="coerce")
```

```python
df.rename(columns={
    "restaurant name":"restaurant_name",
    "restaurant type":"restaurant_type",
    "rate (out of 5)":"rating",
    "num of ratings":"num_ratings",
    "avg cost (two people)":"avg_cost_two_people",
    "table booking":"table_booking",
    "online_order":"online_order",
    "cuisines type":"cuisines_type",
    "local address":"local_address"
}, inplace=True)
```

```python
df["rating_category"] = np.where(
    df["rating"] >= 4,
    "Highly Rated",
    "Average Rated"
)
```

```python
df.head()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
